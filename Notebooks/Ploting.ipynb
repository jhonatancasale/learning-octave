## Naive initializations

```octave
t = [0:0.01:0.98];

y1 = sin(2*pi*4*t);
y2 = cos(2*pi*4*t);
```

## Plot

```octave
plot(t, y1);
```

```octave
plot(t, y2)
```

### Combining many plots at once using `hold on`

```octave
plot(t, y1);
hold on;
plot(t, y2, 'r'); % r for "red" color
xlabel('time')
ylabel('value')
legend('sin', 'cos')
title('Sample plot')

cd '/tmp';
print -dpng 'Sample plot.png';
```

### Using `figure`

```octave
figure(1); plot(t, y1);
```
